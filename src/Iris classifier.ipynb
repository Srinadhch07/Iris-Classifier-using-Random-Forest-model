# Iris Classfication using Random Forest  Classifier

### Importing  libraries 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Adding  column names to CSV file 
column_names=['sepalLengths','sepalWidth','petalLength','petalWidth','species']

#Resulting data in terms of species = 0 1 2 
species=['Setosa','Versicolor','Verginica']

#Loading training data
#header=0 used for steting the 1st row as names of columns
train_data=pd.read_csv('iris_training.csv',names=column_names,header=0)
#print(train_data)


#Loading testing data
test_data=pd.read_csv('iris_test.csv',names=column_names,header=0)
#print(test_data)

#Extracting the attributes of training data
x_train=train_data[['sepalLengths','sepalWidth','petalLength','petalWidth']]
#print(x_train)

#Extracting the attributes of testing data
x_test=test_data[['sepalLengths','sepalWidth','petalLength','petalWidth']]
#print(x_test)


#Extracting species column from trained data
y_train=train_data.pop('species')
#print(y_train)

#Extracting species column from testing data
y_test=test_data.pop('species')
#print(y_test)



#Checking number of rows in trained data
train_data.shape
# Since we can't show all arrtibutes ( x and y must be the same size) so we taking sepallength

plt.scatter(x_train[['sepalLengths']],y_train)
plt.xlabel('Sepal Lengths')
plt.ylabel('Type of Iris')
plt.title('Graphs of training data')
plt.show()



```

## Building the model

```python
model=RandomForestClassifier(n_estimators=100,max_depth=5,max_features=5)
```

### Training the model

```python
model.fit(x_train,y_train)
```

### Testing the model

```python
predictions=model.predict(x_test)

accuracy=accuracy_score(y_test,predictions)

print("The accuracy is :",accuracy)
```

### The accuracy is : 0.9666666666666667

### Model is ready for production.

```python
# sample input
#  5.0         2.3          3.3         1.0
# Expected output is 1 (Versicolor)
values=input().split()
values=list(map(float,values))
predict=model.predict([values])
print(species[int(predict)])
```
